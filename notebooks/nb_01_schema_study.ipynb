```python
# for reading env file
from dotenv import load_dotenv
import os

# for loading db
import pymysql

# for working part
import numpy as np
import pandas as pd

# for config saving
import json
```

```python
load_dotenv()

conn = pymysql.connect(host=os.getenv('host'),
                       port=int(os.getenv('port')),
                       user=os.getenv('user'),
                       password=os.getenv('password'),
                       database=os.getenv('database'),
                       ssl={'ssl_mode':'REQUIRED'})

cursor = conn.cursor()
database_name = os.getenv('database')
print(f'database connected, database name is {database_name}')
```

```python
tables = pd.read_sql('show tables;',conn)

tables
```

```python
selected_tables = ['calls','contacts','leads','order_items','orders','pii','products','order_history']
```

```python
config = {'selected_tables':selected_tables}


with open('../config/selected_tables.json',"w") as f:
    json.dump(config,f,indent=4) 
```

```python
for i in selected_tables:
    df = pd.read_sql(f'select * from {i} limit 5',conn)
    print(i)
    print(df.columns)
    print()
```
